```python
import pandas as pd

data = 'wbc.csv' # path dataset
df = pd.read_csv(data)
df.head(10)
```

```python
df.info()
```

task 1

```python
# Memilih kolom yang diinginkan dari data
df= df[['diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']]

# Menampilkan lima baris pertama dari data yang dipilih
df.head()
```

task 2

```python
# Melakukan pengodean pada kolom "diagnosis"
df['diagnosis_encoded'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Menampilkan lima baris pertama dari data yang telah dienkode
df.head()
```

task 3

```python
# Memilih semua kolom dengan nilai numerik
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Inisialisasi StandardScaler
scaler = StandardScaler()

# Melakukan standarisasi pada kolom-kolom numerik
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Menampilkan lima baris pertama dari data yang telah distandarisasi
df.head()
```

task 4

```python
# Memisahkan fitur dan target
X = df.drop(columns=['diagnosis'])  # Fitur
y = df['diagnosis']  # Target

# Split data training dan data lainnya
# data lainnya akan kita split lagi menjadi validasi dan testing.
# Rasio yang akan kita gunakan adalah 80:20
X_train, X_unseen, y_train, y_unseen = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)

# Split lagi antara validasi dan testing
# Rasio validasi dan testing adalah 1:1 dari data yang belum terlihat
X_val, X_test, y_val, y_test = train_test_split(X_unseen, y_unseen, test_size=0.5, random_state=0, stratify=y_unseen)

# Cek masing-masing ukuran data
print(f'Jumlah data asli: {X.shape[0]}')
print(f'Jumlah data train: {X_train.shape[0]}')
print(f'Jumlah data val: {X_val.shape[0]}')
print(f'Jumlah data test: {X_test.shape[0]}')

# Cek rasio tiap label
print('=========')
print(f'Jumlah label data asli:\n{y.value_counts()}')
print(f'Jumlah label data train:\n{y_train.value_counts()}')
print(f'Jumlah label data val:\n{y_val.value_counts()}')
print(f'Jumlah label data test:\n{y_test.value_counts()}')
```
