# LAB ASSIGNMENT


```python
#Step 1 - Load Data
import pandas as pd
data = 'wbc.csv' # path dataset
df = pd.read_csv(data) # load dataset

df.head()
```

```python
# Separate Variables (Features)
# Check the columns in the dataset
df.info()

# Drop columns that should not be used for modeling, e.g., 'id'
df = df.drop(columns=['id'])  # Assuming 'id' is a non-useful column

# After dropping unnecessary columns, check the remaining ones
df.columns
```

```python
#Perform Encoding on the "diagnosis" Column
# Encoding the "diagnosis" column: M -> 1, B -> 0
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Check the encoding
df['diagnosis'].value_counts()
```

```python
#Standardize All Numeric Columns
from sklearn.preprocessing import StandardScaler

# Identify numeric columns (excluding the target 'diagnosis')
numeric_columns = df.columns.drop('diagnosis')

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the numeric columns
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Check the standardized data
df.head()
```

```python
#Perform a Stratified Split of the Data
from sklearn.model_selection import train_test_split

# Split the data into features (X) and target (y)
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']

# Perform stratified split with an 80:20 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Check the shape of the resulting datasets
(X_train.shape, X_test.shape), (y_train.shape, y_test.shape)
```
