## STEP 3 - Imputation

```python
#Step 3 - Imputation
# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)
```

perform imputation for missing data in "Age," "Cabin," and "Embarked." The strategy we will use is as follows,

"Age" --> Since "Age" is nominal data, we will use the mean strategy.

"Cabin" --> "Cabin" is related to passenger cabin numbers. Here, we will assume that all passengers without cabin numbers stayed on the ship's decks. Therefore, we will substitute missing data with the information "DECK."

"Embarked" --> "Embarked" represents the location of passenger embarkation. There are only 2 missing data points here. Since the data is nominal, we can use the mode to substitute missing data.

# **Lab Work 2**

Feature Extraction
In this stage, we will select which variables we will use as features. In the foundational theory section, we have already learned that not all variables can always be used as features. Therefore, we can choose features that are deemed important for the purpose of creating a machine learning model. We will use the same case, which is to identify the characteristics of Titanic passengers who survived. Thus, the features we will use are,

"Pclass"

"Cabin"

We will separate these variables from the others. The data we will use is the Titanic data that has been repaired through the imputation process.

## Step 0 - Load Library

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

## Step 1 - Load Data

```python
dpath = '/content/Titanic-Dataset-fixed.csv'
df = pd.read_csv(dpath)
df.head()
```

## Step 2 - Slice The Data

```python
#We will select only the columns that we think is features.
df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']]
df.head()
```

## Step 3 - Encoding

We will perform encoding on categorical data, namely "Sex" and "Cabin." This step is necessary because machine learning algorithms perform mathematical calculations. The strategy we will use is Label Encoding. Label Encoding is similar to Ordinal Encoding, with the only difference being the sorting process. Label Encoding does not sort the data beforehand.

```python
le = LabelEncoder() # initiate LabelEncoder object
df['Sex'] = le.fit_transform(df['Sex']) # encoding
df['Cabin'] = le.fit_transform(df['Cabin']) # encoding
```

## Step 4 - Verification

```python
#rerun df.head command to verify the result.
df.head()
```

## Step 5 - Standardization

In this stage, we will standardize "Age." This is not actually necessary because the values in "Age" are not too far apart. The purpose of this step is to demonstrate the standardization process.

```python
std = StandardScaler()
df['Age'] = std.fit_transform(df[['Age']])
```
