Data Imputation : In this work, we will attempt to perform data imputation on missing values.

Step 1 - Load Data

```python
import pandas as pd

data = 'Titanic-Dataset.csv' # path dataset
df = pd.read_csv(data) # load dataset
```

```python
df.head() #show 5 first data
```

Step 2 - Data Checking

```python
df.info() #to  get general information about the data
```

```python
df.isnull().sum() # check the total null data
```

Step 3 - Imputation

```python
# Age - mean
df['Age'].fillna(value=df['Age'].mean(), inplace=True)

# Cabin - "DECK"
df['Cabin'].fillna(value="DECK", inplace=True)

# Embarked - modus
df['Embarked'].fillna(value=df['Embarked'].mode, inplace=True)

#The code is designed to handle missing data by replacing them with meaningful values:
# Age is filled with the mean value.
# Cabin is filled with the string "DECK".
# Embarked is filled with the most frequent value (mode).
```

Step 4 - Verification

Separate the variables that can be used and the ones that cannot be used.

Perform encoding on the "diagnosis" column.

Standardize all columns with numeric values.

Perform a stratified split of the data to create training and test data with an 80:20 ratio.

Step 1 Load data & Library

```python
import pandas as pd

df5 = pd.read_csv('wbc.csv')
df5.head() #check data
```

Step 2 - Slice The Data

```python
df6 = df5.drop(columns=['id', 'Unnamed: 32']) #drop unuseful column

df6.head() #show updated data
```

Step 3 - Encoding

```python
le = LabelEncoder() # initiate LabelEncoder object
df6['diagnosis'] = le.fit_transform(df6['diagnosis']) # encoding diagnosis

df6.head()#show updated data
```

Step 4 - Standardization

```python
std = StandardScaler()
df6[df6.columns] = std.fit_transform(df6)

df6.head() #show updated data
```

Step 5 - Split Data

```python
from sklearn.model_selection import train_test_split

#Perform a stratified split of the data to create training and test data with an 80:20 ratio.
df6_train, df6_test = train_test_split(df6, test_size=0.2, random_state=0, stratify=df6['diagnosis'])

# Check the size of each split
print(f'Number of original data labels:\n{df6.diagnosis.value_counts()}')
print(f'Number of train data labels:\n{df6_train.diagnosis.value_counts()}')
print(f'Number of test data labels:\n{df6_test.diagnosis.value_counts()}')
```
