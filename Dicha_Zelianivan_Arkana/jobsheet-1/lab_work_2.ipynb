Load the data and load the library we're going to need. After loading the data, also show the first 5 rows to get the gist of it.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

path = '/content/drive/MyDrive/Machine Learning/Titanic-Dataset-fixed.csv'
df = pd.read_csv(path)
df.head()
```

Select the columns that we need and consider as features. Which are `survived`, `age`, `sex`, and `cabin`

```python
df = df[['Survived', 'Age', 'Sex', 'Cabin']]
df.head()
```

We'll transform the data, or rather, 'encode' the data to its numerical presentation since machine learning works on mathematical numbers. We can use the label encoder provided by sklearn

```python
le = LabelEncoder() # initiate LabelEncoder object
df['Sex'] = le.fit_transform(df['Sex']) # encoding
df['Cabin'] = le.fit_transform(df['Cabin']) # encoding
```

Let's verify the result using `df.head()`

```python
df.head()
```

After doing that, we need to scale the age to be closer values. We need to standardize them into a certain scale. We can use `StandardScaler`

```python
std = StandardScaler()
df['Age'] = std.fit_transform(df[['Age']])

# verify the result
df.head()
```
