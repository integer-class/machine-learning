From the dataset provided, all variables except id and diagnosis are numeric and can be used for predictive modeling. The id column is simply an identifier and doesn't contribute to the model.

## 2. Perform ecoding on the "diagnosis" column

The diagnosis column contains two categorical values: 'M' (Malignant) and 'B' (Benign). These need to be converted to numeric values using encoding.
A common way to encode these is by mapping 'M' to 1 and 'B' to 0.

```python
# Encode the 'diagnosis' column
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})

```
